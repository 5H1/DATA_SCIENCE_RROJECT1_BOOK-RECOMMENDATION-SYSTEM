```python
#1.LOCATION CATEGORICAL COLUMNS
```

```python
# Calculate frequency of each category
category_counts = merged_df2['Location'].value_counts()
```

```python
category_counts
```

```python
# Get the top 10 categories
top_10_categories = category_counts.head(10)
```

```python
top_10_categories
```

```python
# Sort categories by their frequencies
sorted_categories = top_10_categories.sort_values(ascending=False)
```

```python
# Plotting the bar graph
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sorted_categories.plot(kind='bar')
plt.xlabel('Location')
plt.ylabel('Frequency')
plt.title('Top 10 Locations by Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
